```python
import requests

github_raw_file_url = 'https://raw.githubusercontent.com/aadivik/theresanaiforthat/main/ai_save.html'

response = requests.get(github_raw_file_url)

if response.status_code == 200:
    # Save the content to a local file
    with open('local_file.html', 'w', encoding='utf-8') as file:
        file.write(response.text)
else:
    print(f"Failed to retrieve the file. Status code: {response.status_code}")
```

```python
from bs4 import BeautifulSoup

with open('local_file.html', 'r', encoding='utf-8') as file:
    html = file.read()

soup = BeautifulSoup(html, 'html.parser')

items = soup.find_all('li', class_='li m')

# Custom sorting key function
def sort_key(item):
    saves_element = item.find(class_='saves')
    saves = saves_element.text if saves_element else None
    return -int(saves) if saves is not None else float('inf')

sorted_items = sorted(items, key=sort_key)

for item in sorted_items:
    data_name = item['data-name']
    saves_element = item.find(class_='saves')
    saves = saves_element.text if saves_element else None
    print(f"Data Name: {data_name}, Saves: {saves}")
```
